## Image read (Star)

```python
# Image read
# the simple one channel gray scale image read [0-255]
def LoadImageGrayScaleAuto(img_path):
    """
    load an image and judge if it is grayscale or not, if not will remind and force the image to convert to grayscale
    return as numpy array [height, width] with values in [0, 255] type uint8
    """
    try:
        img = Image.open(img_path)
    except Exception as e:
        raise RuntimeError(f"Failed to load image: {img_path}. Error: {str(e)}")

    width, height = img.size  # Pillow scales images as (width, height)
    print(f"Input image size: {width} x {height} (width x height)")

    original_mode = img.mode
    grayscale_modes = {"1", "L", "LA"}

    if original_mode not in grayscale_modes:
        print(f"Warning: Input image is not grayscale (current mode: {original_mode}). "
              f"Converting to 8-bit grayscale automatically.")
        img_gray = img.convert("L")
    else:
        img_gray = img

    gray_array = np.array(img_gray, dtype=np.uint8)

    return gray_array # return as numpy array 

# Image read color
def LoadImageColorAuto(img_path):
    """
    Load an image and automatically unify it to standard 8-bit RGB format.
    If the input is not standard RGB (e.g. RGBA, CMYK, grayscale, palette),
    it will print a reminder and force conversion.
    Returns:
        numpy.ndarray: RGB image array with shape (height, width, 3),
                       dtype uint8, pixel values in [0, 255], channel order: R, G, B
    """
    try:
        img = Image.open(img_path)
    except Exception as e:
        raise RuntimeError(f"Failed to load image: {img_path}. Error: {str(e)}")

    width, height = img.size # Pillow scales images as (width, height)
    print(f"Input image size: {width} x {height} (width x height)")

    original_mode = img.mode

    if original_mode == "RGB":
        img_rgb = img
    else:
        grayscale_modes = {"1", "L", "LA"}
        if original_mode in grayscale_modes:
            print(f"Warning: Input is a grayscale image (mode: {original_mode}). "
                  f"Converting to 3-channel RGB automatically.")
        else:
            print(f"Warning: Input image is not standard RGB (current mode: {original_mode}). "
                  f"Converting to 8-bit RGB automatically.")
        img_rgb = img.convert("RGB")
    
    rgb_array = np.array(img_rgb, dtype=np.uint8)
    # If use OpenCV, the channel order will be BGR, but here we keep it as RGB for consistency. " bgr_array = rgb_array[:, :, ::-1]  # RGB → BGR "

    return rgb_array

# If you need use SVG image need new function to load and export (cairosvg can convert SVG to PNG or PDF, then you can load it as a normal image)
```

## Data show

```python
# Image data visualization
def VisualizeImage(img_array, title="Image"):
    """
    Visualize an image represented as a numpy array.
    Args:
        img_array (numpy.ndarray): The image data to visualize. Can be grayscale or RGB.
        title (str): Title for the visualization window.
    """
    import matplotlib.pyplot as plt

    if img_array.ndim == 2:  # Grayscale image
        plt.imshow(img_array, cmap='gray', vmin=0, vmax=255)
    elif img_array.ndim == 3 and img_array.shape[2] == 3:  # RGB image
        plt.imshow(img_array)
    else:
        raise ValueError("Unsupported image array shape for visualization.")

    plt.title(title)
    plt.axis('off')  # Hide axis
    plt.show()
```

## Image Jumble

```python
# ImageJumbleTools
# Location change , Pixel change , Field change , Chaos change

def _random_pixel_encrypt(img_array, seed=None):
    # 1. 校验维度并展平像素（保持每个像素的通道完整性）
    if img_array.ndim == 2:
        height, width = img_array.shape
        channels = 1
        flat_pixels = img_array.reshape(-1)  # 形状：(总像素数,)
    elif img_array.ndim == 3:
        height, width, channels = img_array.shape
        flat_pixels = img_array.reshape(-1, channels)  # 形状：(总像素数, 通道数)
    else:
        raise ValueError(
            "Unsupported image dimensions. "
            "Only 2D grayscale (H,W) or 3D RGB (H,W,3) images are supported."
        )

    total_pixels = height * width

    # 2. 生成/记录随机种子（密钥核心）
    if seed is None:
        # 自动生成 32 位整数随机种子
        seed = np.random.randint(0, 2**31 - 1)

    # 3. 独立随机状态：不影响全局随机种子，保证函数可重入、安全
    rng = np.random.RandomState(seed)
    # 生成像素位置的随机排列
    permutation = rng.permutation(total_pixels)

    # 4. 按排列打乱像素位置
    shuffled_flat = flat_pixels[permutation]

    # 5. 恢复原图维度
    if channels == 1:
        jumbled_img = shuffled_flat.reshape(height, width)
    else:
        jumbled_img = shuffled_flat.reshape(height, width, channels)

    # 6. 构造完整密钥（解密所需的全部信息）
    key = {
        "mode": "Locationjumble",
        "height": height,
        "width": width,
        "channels": channels,
        "seed": seed
    }

    return jumbled_img.astype(img_array.dtype), key

def _random_pixel_decrypt(jumbled_array, key):
    # 密钥校验
    if key["mode"] != "random_pixel_shuffle":
        raise ValueError(
            f"Key mode mismatch: expected 'random_pixel_shuffle', got '{key['mode']}'"
        )

    height = key["height"]
    width = key["width"]
    channels = key["channels"]
    seed = key["seed"]
    total_pixels = height * width

    # 输入尺寸校验
    expected_shape = (height, width) if channels == 1 else (height, width, channels)
    if jumbled_array.shape != expected_shape:
        raise ValueError(
            f"Image shape mismatch: expected {expected_shape}, got {jumbled_array.shape}"
        )

    # 1. 用相同种子生成完全一致的排列
    rng = np.random.RandomState(seed)
    permutation = rng.permutation(total_pixels)

    # 2. 计算逆排列（解密核心：argsort 得到还原索引）
    inverse_perm = np.argsort(permutation)

    # 3. 展平混淆图
    if channels == 1:
        flat = jumbled_array.reshape(-1)
    else:
        flat = jumbled_array.reshape(-1, channels)

    # 4. 按逆排列还原像素顺序
    restored_flat = flat[inverse_perm]

    # 5. 恢复原图维度
    if channels == 1:
        restored_img = restored_flat.reshape(height, width)
    else:
        restored_img = restored_flat.reshape(height, width, channels)

    return restored_img.astype(jumbled_array.dtype)


JumbleMode = {"Locationjumble":(_random_pixel_encrypt, _random_pixel_decrypt),
              "Pixeljumble":"random",
              "Fieldjumble":"random",
              "Chaosjumble":"random"

             }


def jumble_image(img_array, mode="Locationjumble", seed=None):
    """
    通用图片混淆加密入口
    
    Args:
        img_array (np.ndarray): 输入图片数组（灰度2D / RGB3D，uint8）
        mode (str): 混淆模式，必须在 JumbleMode 中注册
        seed (int, optional): 自定义随机种子；为 None 时自动生成
    
    Returns:
        tuple: (混淆后图片数组, 解密密钥字典)
    """
    if mode not in JumbleMode:
        available = list(JumbleMode.keys())
        raise ValueError(f"Unsupported mode: {mode}. Available modes: {available}")

    encrypt_func, _ = JumbleMode[mode]
    return encrypt_func(img_array, seed=seed)






```

## Decode Restore

```python
# Restore the original image from a jumbled image using the provided jumble mode.

def unjumble_image(jumbled_array, key):
    """
    通用图片解密还原入口
    
    Args:
        jumbled_array (np.ndarray): 混淆后的图片数组
        key (dict): jumble_image 生成的解密密钥
    
    Returns:
        np.ndarray: 还原后的原始图片数组
    """
    mode = key.get("mode")
    if mode not in JumbleMode:
        raise ValueError(f"Unsupported key mode: {mode}")

    _, decrypt_func = JumbleMode[mode]
    return decrypt_func(jumbled_array, key)
```
